```python
import seaborn as sns 
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.shape
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri)
plt.show() # in seaborn we dont have to give label it take sit automatically and the syntax is prety easy than matlpotlib
```

```python
#  Grouping data with 'hue'
```

```python
sns.lineplot(x="timepoint", y= "signal",data = fmri,hue="event")
plt.show()
```

```python
#  Adding Style
```

```python
sns.lineplot(x="timepoint", y= "signal",data = fmri,hue="event",style="event")
plt.show()
```

```python
# Adding Markers
```

```python
sns.lineplot(x="timepoint", y= "signal",data = fmri,hue="event",style="event",markers = True)
plt.show()
```

```python
# Histogram  :- for this inn seaborn distplot commnd is used
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
diamonds = pd.read_csv('diamonds.csv')
diamonds.head()
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
sns.distplot(diamonds['price'],hist=False)# with hist commnad histofraam is not shown only the line is shown
plt.show()
```

```python
sns.distplot(diamonds['price'],color="red")
plt.show()
```

```python
sns.distplot(diamonds['price'],color="red",kde=False,bins=5)
plt.show() # to remove curve 'kde = False' is used
```

```python
sns.distplot(diamonds['price'],color="orange",kde=False,bins=5,vertical =True)
plt.show()
```

```python
#  SeaBorn JointPlot
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
sns.jointplot(x='SepalLengthCm', y='PetalLengthCm',data=iris)
plt.show()
```

```python
sns.jointplot(x='SepalLengthCm', y='PetalLengthCm',data=iris, color="olive")
plt.show() # used color command for changning the color.
```

```python
sns.jointplot(x='SepalLengthCm', y='PetalLengthCm',data=iris,color="red",kind="reg")
plt.show() # "Kind=reg" to used the regression in the chart command is used.
```

```python
#  Box Plot
```

```python
churn = pd.read_csv('churn.csv')
churn.head()
```

```python
sns.boxplot(x ="Churn", y= "tenure",data= churn)
plt.show()
```

```python
sns.boxplot(x ="InternetService", y= "MonthlyCharges",data= churn)
plt.show()
```

```python
sns.boxplot(x ="InternetService", y= "tenure",data= churn,palette="Set1")
plt.show()
```

```python
sns.boxplot(x ="Contract", y= "tenure",data= churn, linewidth=4)
plt.show() # 'linewidth =' command used to give thiockness to the borders of the plots.
```

```python
#   If you have to change the order then following is thee code.

sns.boxplot(x ="Contract", y= "tenure",data= churn, order=["Two year", "Month-to-month","One year"])
plt.show()
```

```python
#''' Hue command also used for relating the 
#    third data with the two relational chart plot data'''
```

```python
sns.boxplot(x ="Contract",y="tenure",data= churn,hue="PaymentMethod")
plt.show()
```
