```python
## MatPlotLIb


# :- matplotlib is a python library used for data visualization
# :-You can create bar-plots, scatter-plots, histograms and a lot more with matplotlib
```

```python
#  Line Plot

import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=3*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
# Adding title and lables to the plot-chart.

plt.plot(x,y)
plt.title(" Line Plot")
plt.xlabel("x-lable")
plt.ylabel("y-lable")
plt.show()
```

```python
# Changing line Asesthetics

plt.plot(x,y,color='g',linestyle=':',linewidth=3)
plt.show()
```

```python
# Adding two lines in the samep lot

x = np.arange(1,11)
x
```

```python
y1=2*x
y1
```

```python
y2=3*x
y2
```

```python
# Adding two lines in the samep lot


plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
# Line plot
## Adding Sub-plots - in it we get two graph in one screen.
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':', linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle=':', linewidth=2)
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python

###  Bar Plot  ###
```

```python
student = {"Bob":100,"Matt":56,"Sam":27}
names =list(student.keys())
values = list(student.values())

plt.bar(names,values)
plt.show()
```

```python
# Adding title and lables

plt.bar(names,values)
plt.title("Distribution of Studens marks")
plt.xlabel("Names of Students")
plt.ylabel("Marks of Students")
plt.grid(True)
plt.show()
```

```python
## for making Bar plot horizontal
# simple replace 'bar' by 'barh' command

plt.barh(names,values, color='g')
plt.title("Distribution of Studens marks")
plt.xlabel("Names of Students")
plt.ylabel("Marks of Students")
plt.grid(True)
plt.show()
```

```python

#### Scatter Plot

```

```python
# Creating a basic Scatter-plot

x=[10,20,30,40,50,60,70,80,90]
y=[11,20,45,5,8,76,32,12,82]

plt.scatter(x,y)
plt.show()
```

```python
#  Changing Mark Aesthetics

plt.scatter(x,y,marker='*',color='r',s=50)
plt.show()

```

```python
# Adding two markers in the same plot

x=[10,20,30,40,50,60,70,80,90]
a=[11,20,45,5,8,76,32,12,82]
b=[7,18,14,6,90,24,55,28,45,]

plt.scatter(x,a,marker='*',color='r',s=50)
plt.scatter(x,b,marker='.',color='g',s=200)
plt.show()
```

```python
# Adding Subplots 

plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="g",s=100)

plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="b",s=200)
plt.show()
```

```python

### Pie-Chart
```

```python
# Ctreating Data

fruit = ['Apple','Orange','Mango','Guava']
quantity = [76,34,100,29]

plt.pie(quantity,labels =fruit)
plt.show()
```

```python
#  Chnaging Asthetics in Pie-chart

plt.pie(quantity,labels=fruit,autopct='%0.1f%%'
        ,colors=['yellow','grey','blue','black'])
plt.show()
```

```python

## Donut chart

plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show
```
