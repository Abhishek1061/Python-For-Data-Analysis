```python
# Pandas Series Object

# :- Series Object is One-dimensional Labeled array

import pandas as pd
s1=pd.Series([1,2,3,4,5])
s1      #this gives series of given data-set
```

```python
type(s1)
```

```python
#  Changing Index

s1 = pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
s1
```

```python
#  Series Object from Dictionary

# :- You can also create a series object from a dictionary!!

import pandas as pd
pd.Series({'a':10,'b':20,'c':30})
```

```python
# Changinf index position

pd.Series({'a':10,'b':20,'c':30}, index=['b','c','d','a'])
# you can change the index
```

```python
#  Extracting Individual Elements
```

```python
#  Extracting a single Element
s1 = pd.Series([1,2,3,4,5,6,7,8,9])
s1[3]
```

```python
# #  Extracting  Elements from back
s1[-3:]
```

```python
#  Extracting a sequence of Elements
s1[:4]
```

```python

# # Basic Math Operations on Series
```

```python
# Adding a scalar value to Series Elements

s1 = pd.Series([1,2,3,4,5,6,7,8,9])
s1 +5
```

```python
#  Adding Two Series Objects

s1= pd.Series([1,2,3,4,5,6,7,8,9])
s2 =pd.Series([10,20,30,40,50,60,70,80,90])

s1+s2
```
